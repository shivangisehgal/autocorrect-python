```python
import numpy as np
import pandas as pd
import re
#Counter to hold the count of each of the elements present the dict
from collections import Counter as ct
#textdistance to implement Jaccard distance algorithm
from textdistance import Jaccard
```

```python
#List of words
words = []
```

```python
with open('big.txt', 'r') as f:
    data = f.read()
    data=data.lower()
    words = re.findall('\w+',data)
print(f'Total number of words in text file is {len(words)}')
```

```python
#Vocabulary containing all words in the text file
vocab = set(words)
print(f"there are {len(vocab)} distinct words in vocabulary")
```

```python
#Storing the count of each word
frequency = {}  
frequency = ct(words)
#Most commmon words
print(frequency.most_common()[0:5])
```

```python
#Relative frequency of each word or probabilty of occurence of each word
rel_freq = {}     
freq_sum = sum(frequency.values())    
for i in frequency.keys():
    rel_freq[i] = frequency[i]/freq_sum
rel_freq
```

```python
#SpellCheck function
def autocorrect(word):
    word = word.lower()
    if word in vocab:
        return('The spelling is correct!')
    else:
        #Calculating similarity between input word and words from vocabulary using jaccard distance algorithm
        jcd = Jaccard()
        similar = [1 - (jcd.distance(v,word)) for v in frequency.keys()]
        #Creating a dataframe with index, word, probability and similarity
        df = pd.DataFrame.from_dict(rel_freq, orient = 'index').reset_index()
        df = df.rename(columns = {'index':'Word', 0:'Probability'})
        df['Similarity'] = similar
        output = df.sort_values(['Similarity', 'Probability'], ascending = False)
        #Returning the most similar and most probable word(in respective order of preference) and top 5 words according to this criteria
        print("Did you mean: " + output['Word'].iloc[0] + " ?\nHere are the top five picks: ")
        return(output.head())
```

```python
#Enter any word :
autocorrect('sercet')
```
